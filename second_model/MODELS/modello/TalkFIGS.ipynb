```python
from Bio import SeqIO
import numpy as np
from scipy import stats # Import the scipy.stats module
import math as mth
import numpy as np
import matplotlib.pyplot as plt


filename='../phytophthora_infestans_t30-4_1_genes.fasta'
gensizes=[]

for seq_record in SeqIO.parse(filename, "fasta"):
    #print(seq_record.id)
    #print(repr(seq_record.seq)) 
    #print(len(seq_record))
    gensizes.append(len(seq_record))
    #raw_input()
print("DONE")
 
```

```python

smin=min(gensizes)
smax=max(gensizes)
print smin
print smax

i=smin
sn=[]
while(i<smax+10):
    sn.append(i)
    i+=100
    
hsz=np.histogram(gensizes,sn) 
fig, axes=plt.subplots(1, 1, figsize=(20, 8))
axes.hist(gensizes, histtype="stepfilled",bins=250, alpha=0.8, normed=True)

ysn=hsz[0]
xsn=np.delete(hsz[1],[len(hsz[1])-1])


#fig, axes = plt.subplots(1, 1, figsize=(20, 8))  
#axes.plot(xsn, ysn)
#axes.set_ylabel("Freq")
#axes.set_xlabel("Genes Size")

normsum=(sum(ysn)*100)
ps=[]
for i in ysn:
    ps.append(float(i)/normsum)
print ps

fig, axes = plt.subplots(1, 1, figsize=(20, 8))  
axes.plot(xsn, ps)
axes.set_ylabel("P(s)")
axes.set_xlabel("Genes Size")


plt.style.use('bmh')
fig, axes = plt.subplots(1, 1, figsize=(20, 8))  
axes.plot(xsn, ps)
axes.set_ylabel("P(s)")
axes.set_xlabel("Genes Size")
axes.set_xscale("log")
#axes.set_yscale("log")
```

```python
NGPI=len(gensizes)
MU=np.mean(gensizes)
MUZ=float(sum(gensizes))/float(NGPI)
STD=np.std(gensizes)
print NGPI
print MU
print MUZ
print STD

#rvs = stats.lognorm.rvs(np.log(2), loc=0, scale=4, size=250) # Generate some random variates as data
#n, bins, patches = plt.hist(rvs, bins=25, normed=True) # Plot histogram

shape, loc, scale = stats.lognorm.fit(gensizes, floc=0) # Fit a curve to the variates
mu = np.log(scale) # Mean of log(X)
sigma = shape # Standard deviation of log(X)

print("FITTTTTT!")
print shape
print loc
print scale
print("parameters of the lognormal")

M = np.exp(mu) # Geometric mean == median
s = np.exp(sigma) # Geometric standard deviation
mode=np.exp(mu - sigma**2)



yt1=np.random.lognormal(mean=np.log(scale), sigma=shape, size=len(gensizes))
i1=min(yt1)
xsn1=[]
while(i1<max(yt1)+10):
    xsn1.append(i1)
    i1+=100
    
hsz1=np.histogram(yt1,xsn1) 
ysn1=hsz1[0]/(100.0*sum(hsz1[0]))
xsn1=np.delete(hsz1[1],[len(hsz1[1])-1])



yt2=stats.lognorm.rvs(sigma, loc=0, scale=scale, size=len(gensizes))
#print yt2
#raw_input("ratota")
i2=min(yt2)
sn2=[]
while(i2<smax+10):
    sn2.append(i2)
    i2+=100
    
hsz2=np.histogram(yt2,sn2) 
ysn2=hsz2[0]/(100.0*sum(hsz2[0]))
xsn2=np.delete(hsz2[1],[len(hsz2[1])-1])




#################
print M 
print s

median = np.exp(mu)
mean = np.exp(mu + (sigma**2/2)) # Note that mean depends on both M and s
print("Here we are!")
print mode
print mean
print median
print("Bye")
x = np.linspace(smin, smax, num=400)
pdf=stats.lognorm.pdf(x, shape, loc=0, scale=scale)

fig, axes = plt.subplots(1, 1, figsize=(10, 8))
#axes.plot(xsn,ps,linewidth=1)
#axes.plot(xsn1,ysn1,linewidth=2)
#axes.plot(xsn2,ysn2,linewidth=2)
axes.hist(gensizes, histtype="stepfilled",bins=1500, alpha=1.0, normed=True,label='P.Infestansa.fasta')
axes.set_ylabel("Prob(s)",fontsize=40)
axes.set_xlabel("Gene Size s",fontsize=40)
axes.plot(x, stats.lognorm.pdf(x, shape, loc=0, scale=scale), 'r', linewidth=2,label='Log-Normal fit') # Plot fitted curve
plt.vlines(mode, 0, pdf.max(), linestyle=':', label='Mode')
plt.vlines(mean, 0, stats.lognorm.pdf(mean, shape, loc=0, scale=scale), linestyle='--', color='green', label='Mean')
plt.vlines(median, 0, stats.lognorm.pdf(median, shape, loc=0, scale=scale), color='blue', label='Median')
leg=plt.legend(prop={'size':26})
axes.set_xlim([0,8000])
axes.xaxis.set_tick_params(labelsize=30)
axes.yaxis.set_tick_params(labelsize=30)
fig.set_size_inches(18.5,10.5)
fig.patch.set_alpha(0.5)
fig.savefig('pinfestans.png',dpi=400)
fig.savefig('pinfestans.svg',dpi=400)



#fig, axes = plt.subplots(1, 1, figsize=(20, 8))  
#axes.plot(xsn, ps, linewidth=3)
#axes.hist(gensizes, histtype="stepfilled",bins=1500, alpha=0.8, normed=True)
#axes.set_ylabel("P(s)")
#axes.set_xlabel("Genes Size")
#axes.plot(x, stats.lognorm.pdf(x, shape, loc=0, scale=scale), 'r', linewidth=2) # Plot fitted curve
#axes.set_xscale("log")
#axes.set_yscale("log")

#fig, axes = plt.subplots(1, 1, figsize=(20, 8))  
#axes.plot(xsn, ps, linewidth=3)
#axes.set_ylabel("P(s)")
#axes.set_xlabel("Genes Size")
#axes.plot(x, stats.lognorm.pdf(x, shape, loc=0, scale=scale), 'r', linewidth=2) # Plot fitted curve
#axes.set_xscale("log")
#axes.set_yscale("log")
```
